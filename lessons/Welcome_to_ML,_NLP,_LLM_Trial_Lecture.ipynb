# Machine Learning  (ML)

## 1. What is Machine Learning?

**Definition:**

Machine Learning is a branch of Artificial Intelligence where computers learn patterns from data and improve performance without being explicitly programmed.

*Instead of giving rules, we give examples (data) and the system learns the rules by itself.*

**Example:**

	•	Traditional programming: If temperature > 37.5 → fever.
	•	ML: Feed thousands of patient data (temperature, symptoms) → system learns to classify fever automatically.

## 3. How Does Machine Learning Work?

	1.	Data Collection → Collect raw data (images, text, signals, numbers).
	2.	Preprocessing → Clean, normalize, and prepare data.

	7.	Feedback Loop → Continuous learning & improvement.

# Natural Language Processing (NLP)

## 1. What is NLP?
- Definition:
NLP is a subfield of Artificial Intelligence that enables computers to understand, interpret, and generate human language (text or speech).

- Goal: Make machines interact with humans naturally.

- Example:
	•	ChatGPT, Google Translate, voice assistants (Siri, Alexa).

## 2. Relationship with Other AI Disciplines

	•	Machine Learning (ML): NLP uses ML algorithms to learn language patterns from data.
	•	Deep Learning (DL): Advanced NLP tasks often use neural networks (RNN, LSTM, Transformer).
	•	Large Language Models (LLMs): e.g., GPT, BERT, trained on massive text datasets for advanced language understanding.
	•	Data Science & Statistics: NLP relies on text analysis, probability, and statistics.

## 3. Core NLP Tasks
	1.	Text Preprocessing
	  •	Tokenization, stemming/lemmatization, stopword removal, lowercasing.
	2.	Text Classification
	  •	Sentiment analysis, spam detection, topic categorization.
	3.	Named Entity Recognition (NER)
	  •	Identify names, places, dates in text.
	4.	Machine Translation
	  •	Translate text from one language to another.
	5.	Question Answering & Chatbots
	  •	Answer questions, conduct dialogues with humans.
	6.	Text Generation
	  •	Generate coherent sentences, stories, code, or summaries.

## 4. NLP Workflow (Pipeline)

	1.	Data Collection → Gather text/corpus (news, books, chat logs).
	2.	Preprocessing → Clean text, tokenize, normalize.
	3.	Feature Extraction → Bag-of-Words, TF-IDF, Word Embeddings (Word2Vec, GloVe).
	4.	Model Training → Classical ML (SVM, Naive Bayes) or Deep Learning (RNN, Transformer).
	5.	Evaluation → Accuracy, F1-score, BLEU score for translation.
	6.	Deployment → Chatbots, summarizers, translators, search engines.

## 5. Modern NLP Approaches

	•	Transformers: e.g., BERT, GPT → handle long context, parallelize training.
	•	Pretrained Models: Fine-tuning for specific tasks → reduces need for huge datasets.
	•	Prompt Engineering & LLMs: Guide large models with instructions for new tasks.


## 6. Example Applications

	•	Chatbots & Virtual Assistants
	•	Text Summarization & Translation
	•	Sentiment Analysis & Customer Feedback Analysis
	•	Content Generation & Code Generation (GPT)
	•	Search Engines & Recommendation Systems


# Large Language Models (LLMs)

## 1. What is a Large Language Model?

- Definition:
A Large Language Model (LLM) is a type of AI model trained on massive amounts of text data to understand and generate human-like language.

- Key capability: Can perform a wide variety of language tasks without task-specific programming.

- Example: GPT, BERT, LLaMA, Claude.

API

## 2. Relationship with NLP and AI

	•	Subfield of NLP: LLMs are advanced NLP models.
	•	Built on Deep Learning: Typically use Transformer architectures.
	•	Connected to AI disciplines:
	•	ML/DL: Core algorithms and optimization.
	•	Data Science: Text corpus collection, preprocessing, and evaluation.
	•	Cognitive Computing: Models human-like reasoning for language tasks.

## 3. Core Concepts
	1.	Tokenization: Split text into smaller units (words, subwords, or characters).
	2.	Embeddings: Convert tokens into vectors representing semantic meaning.
	3.	Transformer Architecture:
		•	Attention mechanism → captures context across long text sequences.
		•	Encoder (BERT) → understanding tasks
		•	Decoder (GPT) → generation tasks
	4.	Pretraining & Fine-tuning:
		•	Pretraining → learn general language patterns from massive data.
		•	Fine-tuning → specialize for specific tasks (QA, summarization).
	5.	Inference & Prompting: Provide inputs or instructions → model generates output.

# The augmented LLM

## 1. Definition

- An Augmented LLM is an LLM that combines:

	1.	Pretrained language understanding (from the original LLM).
	2.	External knowledge sources or tools (databases, APIs, search engines, documents).

- Goal: Make the model more accurate, up-to-date, and task-specific.

- Example: Instead of only relying on its training data, the LLM can look up the latest Wikipedia articles or company databases to answer questions.


## 2. Why Augmentation?

	•	LLMs alone may:
	•	Hallucinate (generate incorrect info).
	•	Be outdated, as their knowledge is frozen at pretraining.
	•	Struggle with specific domain knowledge (medical, legal, enterprise data).
	•	Augmentation solves this by connecting the model to external data or reasoning tools.

## 3. Common Augmentation Methods

1. **Retrieval-Augmented Generation (RAG)**
	  •	Model queries a knowledge base and conditions its output on retrieved documents.
	  •	Example: Using FAISS or ElasticSearch to find relevant passages.
2. **Tool-Augmented LLMs**
	  •	LLM calls external tools or APIs for calculations, databases, or other tasks.
	  •	Example: Python code execution, calculators, or web search APIs.
3. **Memory-Augmented LLMs**
	  •	Keeps a persistent context or memory across sessions.
	  •	Example: Personal assistant remembering your past questions.
4. **Instruction / Prompt Engineering**
	  •	Provides structured guidance or templates to improve model output reliability.

## 4. Architecture Example

User Prompt → Augmented LLM → Retrieval/Tools → LLM Output

- Step 1: User asks a question.
- Step 2: LLM queries external sources if needed.
- Step 3: LLM generates answer using retrieved data + pretraining knowledge.

## Hands-On Script: Augmented LLM (RAG)

```python
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from langchain_community.vectorstores import FAISS
from langchain.text_splitter import CharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
```

### model_name = "google/flan-t5-base"

```python
# -------------------------------
# 1. Setup Regular LLM (Flan-T5)
# -------------------------------
model_name = "google/flan-t5-base"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

def regular_generate(prompt, max_length=100):
    """Generate text directly from LLM based on prompt."""
    inputs = tokenizer(prompt, return_tensors="pt")
    if torch.cuda.is_available():
        inputs = {k: v.to("cuda") for k, v in inputs.items()}
        model.to("cuda")
    output = model.generate(
        **inputs,
        max_new_tokens=max_length,
        do_sample=True,
        top_p=0.9,
        temperature=0.7 # 0 - 1, 0.5 < temp, temp < 0.5
    )
    return tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
# -------------------------------
# 2. Prepare Knowledge Base
# -------------------------------
docs = [
    "The capital of France is Paris.",
    "Eiffel Tower is a famous landmark in Paris.",
    "The currency of France is the Euro.",
    "The Louvre Museum is located in Paris and houses the Mona Lisa."
]

# Split documents into smaller chunks
text_splitter = CharacterTextSplitter(chunk_size=100, chunk_overlap=0)
chunks = text_splitter.split_text(" ".join(docs))

# Convert chunks to embeddings and create FAISS vector store
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vector_store = FAISS.from_texts(chunks, embeddings)
```

```python
# -------------------------------
# 3. Augmented LLM (RAG)
# -------------------------------
def augmented_qa(question, top_k=4, max_length=100):
    """Performs RAG-based question answering."""
    # Retrieve relevant documents from the knowledge base
    docs = vector_store.similarity_search(question, k=top_k)
    context = "\n".join([d.page_content for d in docs])

    # Construct prompt with explicit instructions
    prompt = f"""
You are a knowledgeable assistant. Use ONLY the context below to answer the question.

Context:
{context}

Question: {question}

Instructions:
- Answer all parts of the question.
- Format output as: "Capital: <capital>, Landmark: <landmark>"
- Separate multiple items using commas or 'and'

Answer:
"""

    print(f"Prompt:\n{prompt}")  # For debugging

    # Generate answer
    inputs = tokenizer(prompt, return_tensors="pt")
    if torch.cuda.is_available():
        inputs = {k: v.to("cuda") for k, v in inputs.items()}
        model.to("cuda")
    output = model.generate(
        **inputs,
        max_new_tokens=max_length,
        do_sample=False,
        top_p=0.95,
        temperature=0.7,
        num_beams=5
    )
    return tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
# -------------------------------
# 4. Demonstrate Regular vs RAG
# -------------------------------
question = "What is the capital of France and what is a famous landmark there?"

print("--- Regular LLM Output ---")
print(regular_generate(question))

print("\n--- Augmented LLM (RAG) Output ---")
print(augmented_qa(question))
```

### model_name = "google/flan-t5-large"

```python
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from langchain_community.vectorstores import FAISS
from langchain.text_splitter import CharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings

# ------------------------
# 1. Setup: Seq2Seq LLM
# ------------------------
model_name = "google/flan-t5-large"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

def generate(prompt, max_length=100):
    """Generate text from the LLM with beam search"""
    inputs = tokenizer(prompt, return_tensors="pt")
    if torch.cuda.is_available():
        inputs = {k: v.to("cuda") for k, v in inputs.items()}
        model.to("cuda")
    output = model.generate(
        **inputs,
        max_new_tokens=max_length,
        temperature=0.7,
        num_beams=5,
        early_stopping=True
    )
    return tokenizer.decode(output[0], skip_special_tokens=True)

# ------------------------
# 2. Prepare Knowledge Base
# ------------------------
docs = [
    "The capital of France is Paris.",
    "Eiffel Tower is a famous landmark in Paris.",
    "The currency of France is the Euro.",
    "The Louvre Museum is located in Paris and houses the Mona Lisa."
]

# Split documents into smaller chunks
text_splitter = CharacterTextSplitter(chunk_size=100, chunk_overlap=0)
chunks = text_splitter.split_text(" ".join(docs))

# Convert chunks to embeddings and create a FAISS vector store
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vector_store = FAISS.from_texts(chunks, embeddings)

# ------------------------
# 3. RAG: Augmented QA
# ------------------------
def augmented_qa(question, top_k=2):
    """RAG-based question answering using seq2seq LLM"""
    # Retrieve top-k relevant chunks
    docs = vector_store.similarity_search(question, k=top_k)
    context = "\n".join([d.page_content for d in docs])

    # Construct clear, structured prompt
    prompt = f"""
You are a knowledgeable assistant. Use ONLY the context below to answer the question.
Answer ALL parts of the question.
Provide each answer on a separate line as follows:
Capital: <capital>
Landmark: <landmark>

Context:
{context}

Question:
{question}

Answer:
"""
    print(f"Prompt:\n{prompt}")  # Optional: inspect prompt
    return generate(prompt, max_length=100)

# ------------------------
# 4. Demonstration
# ------------------------
question = "What is the capital of France and what is a famous landmark there?"

print("--- Regular LLM Output ---")
print(generate(question))

print("\n--- Augmented LLM (RAG) Output ---")
print(augmented_qa(question))
```
